### 1) Prime Numbers

```python
'''
Given an integer list, return a list containing only prime numbers
'''
def prime(n):
    if n < 2: return False
    else:
        for i in range(2,n):
            if n % i == 0: return False
        return True

try:
    A = [int(i) for i in input('Enter a set of numbers: ').split()]
    P = [v for v in A if prime(v)]
    print('Prime Numbers:', P)
except EOFError: print('No input detected...')
except ValueError: print('Invalid Input...')
```

### 2) Find Common Characters

```python
'''
Returns a string of the characters (excluding whitespace) strings msg1 and msg2 have in common
'''
def find_common_characters(msg1,msg2):
    S = set()
    for c in msg1:
        if c in msg2: S.add(c)
    return ''.join(S)

try:
    A = input('Sentence A: ').replace(' ','')
    B = input('Sentence B: ').replace(' ','')

    m = find_common_characters(A,B)
    if m: print(f'\nCharacters in common: {m}')
    else: print('\n-1')
except EOFError: print('Not enough inputs...')
```

### 3) Pairs of Numbers

```python
'''
Given a list of numbers and an integer n, output the number of integer pairs that yield a sum of n (Note: both
elements in the pair must be unique)
'''
def find_pairs_of_numbers(num_list,n):
    P = set()
    for i in num_list:
        if n-i in num_list and i!=n-i: P.add((min(i,n-i),max(i,n-i)))
    return len(P)

try:
    A = set([int(i) for i in input('Enter a set of numbers: ').split()])
    n = int(input('Enter a number: '))
    p = find_pairs_of_numbers(A,n)
    print(f'\n{p} number pair(s) with a sum of {n}')
except EOFError: print('Not enough inputs...')
except ValueError: print(f'Invalid input...')
```
