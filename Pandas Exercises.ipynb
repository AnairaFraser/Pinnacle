```python
import pandas as pd
from numpy import average as avg
```

<div align="justify"><b>1) Import the data into Python environment as a Pandas DataFrame.</b></div>

```python
try:
    rain = pd.read_csv('rainfall.csv')
    display(rain)
except FileNotFoundError: print('File not found...')
```

<div align="justify"><b>2) Check for missing values, if any and drop the corresponding rows.</b></div>

```python
rain.info()
```

As shown above, there are no missing values in the dataset. This is further demonstrated by the fact that, when implementing the method that removes entries with missing values, the dataset remains unchanged:

```python
rain = rain.dropna()
display(rain)
```

<div align="justify"><b>3) Find the district that gets the highest annual rainfall.</b></div>

```python
hi = rain[rain['ANNUAL'] == max(rain['ANNUAL'])]['DISTRICT'].values[0]
print('District with highest annual rainfall:',hi)
```

<div align="justify"><b>4) Display the top 5 states that get the highest annual rainfall.</b></div>

```python
top5 = rain.sort_values('ANNUAL', ascending = False)
display(top5.head())
```

<div align="justify"><b>5) Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'.</b></div>

```python
['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']
raindrop = rain.drop(['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'], axis = 1)
display(raindrop)
```

<div align="justify"><b>6) Display the state-wise mean rainfall for all the months using a pivot table.</b></div>

```python
piv = pd.pivot_table(raindrop, index = 'STATE_UT_NAME', aggfunc = avg).round(3)
display(piv)
```

<div align="justify"><b>7) Display the count of districts in each state.</b></div>

```python
states = sorted(set(rain['STATE_UT_NAME']))
```

```python
for s in states:
    d = rain[rain['STATE_UT_NAME'] == s].shape[0]
    print(f'Number of States in {s}: {d}')
```

<div align="justify"><b> 8) For each state, display the district that gets the highest rainfall in May. Also display the recorded rainfall.</b></div>

```python
for s in states:
    df = rain[rain['STATE_UT_NAME'] == s]
    m = df[df['MAY'] == max(df['MAY'])]
    d = m['DISTRICT'].values[0]
    r = max(df['MAY'])
    print(f'State: {s}\nDistrict: {d}\nRainfall: {r}\n')
```
