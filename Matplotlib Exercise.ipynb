```python
import matplotlib.pyplot as plt
from IPython.core.display import HTML
from numpy import arange,cos,pi,exp
```

```python
HTML('''
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
''')
```

```python
X = arange(0,5,0.1)
Y = cos(2*pi*X)*exp(-X)

f = plt.figure(figsize = [8,6])
plt.scatter(X,Y, marker = 'o',c = 'c')
plt.plot(X,Y,c='k')
plt.annotate('2nd crest', xy=(2, 0.15), xytext=(2.5, 0.3), arrowprops= dict(fc='green', shrink=.15))
plt.annotate(' = 60', xy=(1, 0.38), xytext=(1.1, 0.41))
plt.xlabel('Abscissa', color = 'g', size = 14)
plt.ylabel('Ordinate', color = 'g', size = 14)
plt.suptitle('Main Heading', size = 18, fontweight = 'bold')
plt.title('Figure Title', size = 16)
plt.grid()
plt.show()
```
